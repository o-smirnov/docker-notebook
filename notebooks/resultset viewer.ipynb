```python
all_dirs = [];
for root, _, files in os.walk(datafolder):
    all_files = []
    for file_ in files:
        if file_.startswith('.'):
            continue
        full_path = os.path.join(root, file_)
        size = os.path.getsize(full_path)
        mtime = os.path.getmtime(full_path);
        all_files.append((mtime,os.path.splitext(file_),size));
    all_files = [ (time.ctime(mtime),file_,ext,size) for mtime,(file_,ext),size in sorted(all_files) ]
    if all_files:
      all_dirs.append((root,DataFrame(all_files,columns=('modified','file','type','size'))))
for name,ds in sorted(all_dirs):
    print "Contents of",name
    display(d)
```

```python
for name, _, _ in all_files:
    if name.lower()[-3:] in ('png', 'jpg', 'jpeg'):
        display(Image(name, width=500))
```
